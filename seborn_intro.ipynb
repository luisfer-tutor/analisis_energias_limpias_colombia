```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Cargar un dataset de ejemplo
df = sns.load_dataset('tips')  # Dataset sobre propinas
print(df.head())  # Inspeccionar el dataset
```

```python
# Gráfico de dispersión entre total_bill y tip
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', style='sex')
plt.title("Relación entre Total Bill y Tip")
plt.show()
```

```python
# Gráfico de barras que muestra el promedio de propinas por día
sns.barplot(data=df, x='day', y='tip', hue='sex', ci=None)
plt.title("Promedio de Propinas por Día y Género")
plt.show()
```

```python
# Gráfico de cajas para analizar la distribución de las cuentas totales por día
sns.boxplot(data=df, x='day', y='total_bill', hue='sex')
plt.title("Distribución de Total Bill por Día y Género")
plt.show()
```

```python
# Histograma de la variable total_bill
sns.histplot(data=df, x='total_bill', kde=True, color='blue', bins=20)
plt.title("Distribución de Total Bill")
plt.show()
```

```python
sns.pairplot(data=df, hue='sex', palette='husl')
plt.title("Gráfico de Pares")
plt.show()
```
